```python
num_ports = len(adaptors)
dp_solutions = [1]

def valid_step(x, y):
    if (adaptors[x] - adaptors[y] <= 3):
        return dp_solutions[y]
    return 0

for x in range(0, num_ports):
    for y in range(0, x):
        if len(dp_solutions) <= x:
            dp_solutions.append(0)
        dp_solutions[x] += valid_step(x, y)
        print(dp_solutions)
            
dp_solutions[-1]
```

```python
def powerset(seq):
    """
    Returns all the subsets of this set. This is a generator.
    """
    if len(seq) <= 1:
        yield seq
        yield []
    else:
        for item in powerset(seq[1:]):
            yield [seq[0]]+item
            yield item
```
