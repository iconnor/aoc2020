```python
import numpy as np
with open("input.txt", "r") as f:
    inp = f.read()

tiles_inp = inp.split("\n\n")
tiles = {}

inner_tiles = 0
for tile in tiles_inp:
    id = int(tile.split(':\n')[0].split()[1])
    arr = []
    for line in tile.split(':\n')[1].splitlines():
        arr.append([1 if c == '#' else 0 for c in list(line)])
    arr = np.array(arr)
    tiles[id] = arr
    inner_tiles += sum(sum(arr[1:-1, 1:-1]))

def get_edges(intiles):
    edges = {}
    for id in intiles.keys():
        arr = intiles[id]
        # N, E, S, W edges
        edges[id] = [arr[0,:], arr[0:,-1], arr[-1,:], arr[0:,0]]
    return edges
edges = get_edges(tiles)
inner_tiles
```

```python
# N - S, E - W, etc
corner_peices = [
    [3,0],[0,1],[1,2],[2,3]
]
import copy
from collections import defaultdict
# Puzzles start with corners
def find_corners(edges, only_two=True):
    unmatched_edges = copy.deepcopy(edges)
    matched_edges = defaultdict(lambda: [None]*4)
    for id in edges.keys():
        for id_i in edges.keys():
            if id == id_i:
                break
            for i in range(4):
                for ii in range(4):
                    A = edges[id][i]
                    B = edges[id_i][ii]
                    if (A == B).all():
                        
                        matched_edges[id][i] = id_i
                        matched_edges[id_i][ii] = id
                        unmatched_edges[id][i] = None
                        unmatched_edges[id_i][ii] = None
                    if only_two and (A == B[::-1]).all():
                        matched_edges[id][i] = id_i
                        matched_edges[id_i][ii] = id
                        unmatched_edges[id][i] = None
                        unmatched_edges[id_i][ii] = None
    return unmatched_edges, matched_edges

def get_sides(edges, only_two=True):
    unmatched_edges, matched_edges = find_corners(edges, only_two)
    ids = []
    for id in matched_edges.keys():
        if only_two == True:
            if sum([x is None for x in matched_edges[id]]) == 2:
                ids.append(id)
        else:
            if sum([x is None for x in matched_edges[id]]) >= 2:
                ids.append(id)
    return ids, unmatched_edges, matched_edges

# 1951    2311    3079
# 2729    1427    2473
# 2971    1489    1171
```

```python
edges = get_edges(tiles)
corners, unmatched_edges, matched_edges = get_sides(edges)
top_left = matched_edges[corners[0]]
print(top_left)
while not (top_left[3] is None and top_left[0] is None):
    print('rotate')
    tiles[corners[0]] = np.rot90(tiles[corners[0]])
    edges = get_edges(tiles)
    corners, unmatched_edges, matched_edges = get_sides(edges)
    top_left = matched_edges[corners[0]]
    print(top_left)

edges = get_edges(tiles)
corners, unmatched_edges, matched_edges = get_sides(edges, only_two=False)
print(corners)
top_left


for id in corners:
    for r in [1,2,3,0]:
        tiles[id] = np.rot90(tiles[id])
        edges = get_edges(tiles)
        new_corners, unmatched_edges, matched_edges = get_sides(edges, only_two=False)
        if len(new_corners) < len(corners):
            corners = new_corners
            print(len(corners), '->', len(new_corners))
            break

corners
```
